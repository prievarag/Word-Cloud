```python
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud

text = "Forecast"

x, y = np.ogrid[:300, :300]

mask = (x - 150) ** 2 + (y - 150) ** 2 > 130 ** 2
mask = 255 * mask.astype(int)


wc = WordCloud(background_color="white", repeat=True, mask=mask)
wc.generate(text)

plt.axis("off")
plt.imshow(wc, interpolation="bilinear")
plt.show()
```

```python
my_list=["Forecast", "Forecast","Forecast","Forecast","Forecast","Forecast","Forecast","Forecast","Forecast","Forecast","Forecast","Forecast","Forecast",
         "Bázis", "Bázis","Bázis","Bázis","Bázis","Bázis","Bázis","Bázis","Bázis","Bázis",
         "Terv","Terv","Terv","Terv","Terv","Terv",
         "Költségek","Költségek","Költségek","Költségek","Költségek","Költségek","Költségek",
         "Risks","Risks","Risks","Risks",
         "Opps","Opps","Opps","Opps","Opps","Opps","Opps",
         "Költséggazda","Költséggazda","Költséggazda","Költséggazda",
         "Eltérés","Eltérés","Eltérés","Eltérés","Eltérés",
         "Indoklás","Indoklás","Indoklás","Indoklás","Indoklás","Indoklás","Indoklás","Indoklás","Indoklás",
         "EBITDA", "EBITDA","EBITDA","EBITDA", "EBITDA","EBITDA","EBITDA", "EBITDA","EBITDA","EBITDA", "EBITDA","EBITDA",
         "Zoom-in", "Zoom-in","Zoom-in","Zoom-in",
         "Zoom-out","Zoom-out","Zoom-out","Zoom-out",
]


#convert it to dictionary with values and its occurences
from collections import Counter
word_could_dict=Counter(my_list)
wordcloud = WordCloud(background_color="white", width = 1000, height = 500).generate_from_frequencies(word_could_dict)

plt.figure(figsize=(15,8))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
#plt.savefig('yourfile.png', bbox_inches='tight')
plt.close()
```
